```python
%matplotlib inline
import pandas as pd
import numpy as np
import keras.backend as K
import keras
```

```python
result_path1 = "planet_predict_2_1_t_1.csv"
result_path2 = "planet_predict_2_3_t_1.csv"
result_path3 = "planet_predict_2_4_t_1.csv"
predict1 = pd.read_csv(result_path1)
predict2 = pd.read_csv(result_path2)
predict3 = pd.read_csv(result_path3)
predict1.head()
```

```python
label = list(predict1.columns)[1:]
print(label)
```

```python
images_name = predict1['name']
images_name = np.array(images_name)
```

```python
images_name
```

```python
del predict1['name']
del predict2['name']
del predict3['name']
y_pred1 = pd.DataFrame.as_matrix(predict1)
y_pred2 = pd.DataFrame.as_matrix(predict2)
y_pred3 = pd.DataFrame.as_matrix(predict3)
y_pred = (y_pred1 + y_pred2 + y_pred3)/3
```

```python
threshhold = 0.3
```

```python
result = []
for i in range(predict1.shape[0]):
    res = ''
    for j in range(len(label)):
        if y_pred[i][j] >= threshhold:
            res = res + label[j] + ' '
    if res:
        res = res[:-1]
    print(res)
    result.append(res)
```

```python
with open('submission2-2.csv', 'w') as outfn:
    s = 'image_name'
    s += ',{}'.format('tags')
    outfn.write(s)
    outfn.write('\n')
    for idx in range(len(images_name)):
        s = images_name[idx]
        s += ',{}'.format(result[idx])
        outfn.write(s)
        outfn.write('\n')
```
